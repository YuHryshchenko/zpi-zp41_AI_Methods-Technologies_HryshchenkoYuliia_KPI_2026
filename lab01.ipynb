<h3>Дослідження способів формування нечітких множин і операцій над ними</h3>

```python
import numpy as np # робота з масивами та числовими векторами
import skfuzzy as fuzz # бібліотека нечіткої логіки
import matplotlib.pyplot as plt # побудова графіків

# Увімкнення сітки на всіх графіках
plt.rcParams['axes.grid'] = True

# ==========================================
# Task 1: Triangular and Trapezoidal MF
# ==========================================
print("Generating Task 1: Triangular and Trapezoidal functions...") # Виводить повідомлення про початок завдання 1

# Створює 100 рівномірних точок у діапазоні від 0 до 10
x1 = np.linspace(0, 10, 100)

# Визначення фукнцій приналежності
triMF = fuzz.trimf(x1, [2, 4, 9]) 
# Трикутна функція приналежності: 0 при x=2; максимум 1 при x=4; 0 при x=9

trapMF = fuzz.trapmf(x1, [1, 3, 5, 7]) 
# Трапецієподібна функція: зростає від 1 до 3; плато від 3 до 5; спадає від 5 до 7

# Побудова графіків
fig1, axes1 = plt.subplots(1, 2, figsize=(16, 6)) 
# Створення одного вікна з графіками (figure), в якому буде 1 ряд і 2 колонки графіків, figsize задає розмір вікна

axes1[0].plot(x1, triMF) # Будує графік функцій приналежності Triangular
axes1[1].plot(x1, trapMF) # Будує графік функцій приналежності Trapezoidal

axes1[0].set_title('Triangular MF') # Надає назву графіку для Трикутної функції
axes1[1].set_title('Trapezoid MF') # Надає назву графіку для Трапецієподібної функції
plt.show() # Зображує графіки 

# ==========================================
# Task 2: Gaussian MF (Simple and Two-sided)
# ==========================================
print("Generating Task 2: Gaussian functions...") # Виводить повідомлення про початок завдання 2

# Створює 100 рівномірних точок у діапазоні від -40 до 40
x2 = np.linspace(-40, 40, 100) 

# Проста функція приналежності Гауса
gauss = fuzz.gaussmf(x2, 13, 7) # Гаусова функція з: середнім значенням = 13 та σ = 7

# Двостороння функція приналежності Гауса
gauss21 = fuzz.gauss2mf(x2, -4, 1, 6, 3) 
# Гаусова двостороння функція: ліва частина — центр -4, σ=1; права частина — центр 6, σ=3
gauss22 = fuzz.gauss2mf(x2, 13, 6, 23, 5) 
# Гаусова двостороння функція: ліва частина — центр 13, σ=6; права частина — центр 23, σ=5
gauss23 = fuzz.gauss2mf(x2, 4, 4, 6, 4) 
# Гаусова двостороння функція: ліва частина — центр 4, σ=4; права частина — центр 6, σ=4

# Побудова графіків
fig2, axes2 = plt.subplots(1, 2, figsize=(16, 6)) 
# Створення одного вікна з графіками (figure), в якому буде 1 ряд і 2 колонки графіків, figsize задає розмір вікна

axes2[0].plot(x2, gauss) # Будує графіки функцій приналежності Gaussian Simple
axes2[0].set_title('Gaussian MF') # Надає назву графіку

axes2[1].plot(x2, gauss21, label="[-4, 1, 6, 3]") # Будує графік функцій приналежності Gaussian Two-sided
axes2[1].plot(x2, gauss22, label="[13, 6, 23, 5]") # Будує графік функцій приналежності Gaussian Two-sided
axes2[1].plot(x2, gauss23, label="[4, 4, 6, 4]") # Будує графік функцій приналежності Gaussian Two-sided
axes2[1].set_title('Gaussian two-combined MF') # Надає назву графіку
axes2[1].legend() # увімкнення пояснювальних підписів для всіх кривих на цьому графіку
plt.show() # Зображує графіки

# ==========================================
# Task 3: Generalized Bell MF
# ==========================================
print("Generating Task 3: Generalized Bell function...") # Виводить повідомлення про початок завдання 3

# Створює 100 рівномірних точок у діапазоні від -5 до 15
x3 = np.linspace(-5, 15, 100)

# Функція приналежності "узагальнений дзвін", де a=3 (ширина), b=4 (крутизна), c=6 (центр)
bell = fuzz.gbellmf(x3, 3, 4, 6) 

# Побудова графіка
fig3, ax3 = plt.subplots() # Створення одного вікна з побудови графіка
ax3.plot(x3, bell) # Будує графік функцій приналежності gbellmf (generalized bell shape membership function)
ax3.set_title("Generalized Bell MF") # Надає назву графіку
plt.show() # Зображує графіки 

# ==========================================
# Task 4: Sigmoid Functions
# ==========================================
print("Generating Task 4: Sigmoid functions...") # Виводить повідомлення про початок завдання 4

# Основна Одностороння сигмоїдна функція
oneSide = fuzz.sigmf(x3, 3, 4) # S-подібна крива
# Додаткова Двостороння сигмоїдна функція
twoSide = fuzz.dsigmf(x3, 3, 4, 7, 8) # Різниця двох сигмоїд → дзвоноподібна форма
# Додаткова Асиметрична сигмоїдна функція
addAsym = fuzz.psigmf(x3, 6, 8, 5, 6) # Добуток двох сигмоїд → асиметрична форма

# Побудова графіків
fig4, axes4 = plt.subplots(1, 3, figsize=(16, 6)) 
# Створення одного вікна з графіками (figure), в якому буде 1 ряд і 3 колонки графіків, figsize задає розмір вікна

axes4[0].plot(x3, oneSide) # Будує графік функцій приналежності Основна Одностороння сигмоїдна функція
axes4[1].plot(x3, twoSide) # Будує графік функцій приналежності Додаткова Двостороння сигмоїдна функція
axes4[2].plot(x3, addAsym) # Будує графік функцій приналежності Додаткова Несиметрична сигмоїдна функція

axes4[0].set_title("Basic One-sided (sigmf)") # Надає назву графіку
axes4[1].set_title("Additional Two-sided (dsigmf)") # Надає назву графіку
axes4[2].set_title("Additional Asymmetric (psigmf)") # Надає назву графіку
plt.show() # Зображує графіки

# ==========================================
# Task 5: Polynomial Functions (Z, PI, S)
# ==========================================
print("Generating Task 5: Polynomial functions...") # Виводить повідомлення про початок завдання 5

z = fuzz.zmf(x3, 3, 4) # поліноміальна крива, що називається Z-функції (zmf)
pi = fuzz.pimf(x3, 3, 4, 5, 6) # поліноміальна крива, що називається РІ-функції (pimf)
s = fuzz.smf(x3, 5, 6) # поліноміальна крива, що називається S-функції (smf)

# Побудова графіків
fig5, axes5 = plt.subplots(1, 3, figsize=(16, 6)) 
# Створення одного вікна з графіками (figure), в якому буде 1 ряд і 3 колонки графіків, figsize задає розмір вікна

axes5[0].plot(x3, z) # Будує графік Z-функції
axes5[1].plot(x3, pi) # Будує графік РІ-функції
axes5[2].plot(x3, s) # Будує графік S-функції

axes5[0].set_title("Z-function") # Надає назву графіку
axes5[1].set_title("Pi-function") # Надає назву графіку
axes5[2].set_title("S-function") # Надає назву графіку
plt.show() # Зображує графіки

# ==========================================
# Task 6: Min-Max Interpretation (Logic Ops) / Мінімаксна інтерпретація
# ==========================================
print("Generating Task 6: Min-Max Interpretation...") # Виводить повідомлення про початок завдання 6

# Використовується змінна x1 (масив значень), визначена раніше в Task 1
# Визначення двох гаусових функцій для роботи із ними
f1 = fuzz.gaussmf(x1, 6, 3) # "Variable A", нечітка множини A
# Створює нечітку множину f1 з гаусовою функцією належності: 
# x1 — область визначення, 6 — середнє значення (центр), 3 — стандартне відхилення (ширина)

f2 = fuzz.gaussmf(x1, 7, 3) # "Variable B", нечітка множини B
# Створює другу нечітку множину f2 з іншим центром - 7 

# Intersection (AND) = Minimum, кон'юнктивний оператор
fmin = np.fmin(f1, f2)
# Обчислює перетин двох нечітких множин: для кожного x береться мінімум між f1 і f2 (логічне AND)

# Union (OR) = Maximum, диз'юнктивний оператор
fmax = np.fmax(f1, f2)
# Обчислює об’єднання двох нечітких множин: Для кожного x береться максимум між f1 і f2 (логічне OR)

# Побудова графіків
fig6, axes6 = plt.subplots(1, 2, figsize=(16, 6)) 
# Створення одного вікна з графіками (figure), в якому буде 1 ряд і 2 колонки графіків, figsize задає розмір вікна

# Plot AND (Min)
axes6[0].plot(x1, f1, label="f1") # Будує графік функції f1 на першому графіку
axes6[0].plot(x1, f2, label="f2") # Будує графік функції f2 на першому графіку
axes6[0].plot(x1, fmin, "r+", label="min") # Будує графік мінімуму (AND) червоними плюсами
axes6[0].set_title("Min Interpretation (AND)") # Надає назву першому графіку
axes6[0].legend() 
# Відображає маленьке віконце на графіку, де написано, як позначені фукнції f1, f2, fmin на першому графіку 

# Plot OR (Max)
axes6[1].plot(x1, f1, label="f1") # Будує графік функції f1 на другому графіку
axes6[1].plot(x1, f2, label="f2") # Будує графік функції f2 на другому графіку
axes6[1].plot(x1, fmax, "r+", label="max") # Будує графік максимуму (OR) червоними плюсами
axes6[1].set_title("Max Interpretation (OR)") # Надає назву другому графіку
axes6[1].legend() 
# Відображає маленьке віконце на графіку, де написано, як позначені фукнції f1, f2, fmax на другому графіку

plt.show() # Зображує графіки

# ==========================================
# Task 7: Probabilistic Interpretation /Імовірнісна інтерпретація кон'юнктивних і диз'юнктивних операторів.
# ==========================================
print("Generating Task 7: Probabilistic Interpretation...") # Виводить повідомлення про початок завдання 7

# Conjunction (AND) = Product; Реалізує ймовірнісну кон'юнкцію (AND) як добуток значень функцій належності
andF = f1 * f2 # Ймовірнісне AND

# Disjunction (OR) = Probabilistic Sum; 
# Реалізує ймовірнісну диз'юнкцію (OR) за формулою: P(A ∪ B) = P(A) + P(B) − P(A ∩ B)
orF = f1 + f2 - andF # Ймовірнісне OR

# Побудова графіків
fig7, axes7 = plt.subplots(1, 2, figsize=(16, 6)) 
# Створення одного вікна з графіками (figure), в якому буде 1 ряд і 2 колонки графіків, figsize задає розмір вікна

axes7[0].plot(x1, f1) # Будує графік функції f1 на першому графіку
axes7[0].plot(x1, f2) # Будує графік функції f2 на першому графіку
axes7[0].plot(x1, andF, "r+", label="Product") # Будує графік мінімуму (AND) червоними плюсами
axes7[0].set_title("Conjunction (Product)") # Надає назву першому графіку

axes7[1].plot(x1, f1) # Будує графік функції f1 на другому графіку
axes7[1].plot(x1, f2) # Будує графік функції f2 на другому графіку
axes7[1].plot(x1, orF, "r+", label="Prob Sum") # Будує графік максимуму (OR) червоними плюсами
axes7[1].set_title("Disjunction (Probabilistic Sum)") # Надає назву другому графіку

plt.show() # Зображує графіки

# ==========================================
# Task 8: Complement (Negation) / математичне уявлення вербального вираження «НЕ А»
# ==========================================
print("Generating Task 8: Complement (Negation)...") # Виводить повідомлення про початок завдання 8

# Complement = 1 - membership; Обчислює доповнення нечіткої множини: для кожного x: μ(¬A) = 1 − μ(A)
notF = 1 - f1

# Побудова графіків
fig8, ax8 = plt.subplots(figsize=(16, 6)) # Створює одне графічне вікно, figsize задає розмір вікна
ax8.plot(x1, f1, label="Function (f1)") # Будує графік початкової функції належності
ax8.plot(x1, notF, "--", label="Complement (NOT f1)") # Будує графік доповнення пунктирною лінією
ax8.set_title("Fuzzy Complement") # Надає назву графіку
ax8.legend() # Відображає маленьке віконце на графіку, де написано, як позначені фукнції f1, notf на графіку 
plt.show() # Зображує графіки

print("All tasks completed.")

# ==============================================================================
# КОНТРОЛЬНІ ЗАПИТАННЯ:
# ==============================================================================
# 1. Нечітка множина — це сукупність елементів, де кожен має ступінь приналежності від 0 до 1.
#    На відміну від чіткої (де приналежність тільки 0 або 1), вона дозволяє описувати розмиті поняття.
#
# 2. Функція приналежності (ФП) математично описує ступінь того, наскільки елемент універсуму 
#    відповідає певному лінгвістичному поняттю.
#
# 3. Значення ФП приймаються в діапазоні [0, 1].
#
# 4. Повна належність означає μ(x) = 1 (елемент точно входить до множини). 
#    Часткова означає 0 < μ(x) < 1 (елемент лише певною мірою відповідає опису).
#
# 5. Приклади: керування клімат-контролем, оцінка ризиків кредитування, 
#    діагностика хвороб, керування двигунами (ABS).
#
# 6. Універсальна множина (універсум) — область визначення, що містить усі можливі значення змінної.
#
# 7. Типові форми: трикутна, трапецієподібна, гауссова, сигмоїдальна, дзвоноподібна.
#
# 8. Трикутна має одну точку максимуму, а трапецієподібна має плато (інтервал значень), де μ(x)=1.
#
# 9. μ(x) = 0.7 означає, що елемент x відповідає властивостям множини на 70%.
#
# 10. α-рівень (α-cut) — це чітка множина всіх елементів, для яких ступінь приналежності μ(x) ≥ α.
#
# 11. Комплементарна множина (НЕ А) визначається як μ(x)_NOT = 1 - μ(x)_A.
#
# 12. Об'єднання (OR) зазвичай використовує операцію max, а перетин (AND) — операцію min.
#
# 13. Нечітке відношення — це нечітка множина на декартовому добутку універсумів, 
#     яка описує силу зв'язку між парами елементів.
#
# 14. ФП використовуються для фазифікації (перетворення чисел у нечіткі терміни) 
#     та дефазифікації (отримання чіткого рішення на виході системи).
#
# 15. Переваги: можливість моделювання складних систем без точних рівнянь, 
#     врахування людського фактору та неточності вхідних даних.
# ==============================================================================
```
